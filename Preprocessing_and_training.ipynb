```python
#The next step is addressing our target variable, Churn. 
#Currently, the values of this feature are “Yes” and “No”. 
#This is a binary outcome, which is what we want, but our model will not be able to meaningfully interpret this in its current string-form. 
#Instead, we want to replace these variables with numeric binary values:
```

```python
df2.Churn.replace({"Yes":1, "No":0}, inplace = True)
```

## Creat dummy features for object type features

```python
dummy_df = pd.get_dummies(df2, drop_first=True)
# dummy_df = pd.get_dummies(df2)
```

```python
# drop outliners
from scipy import stats
dummy_df = dummy_df[(np.abs(stats.zscore(dummy_df)) < 3).all(axis=1)]
```
